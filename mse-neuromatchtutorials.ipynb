<p style="font-size:20px;">
In this exercise, we compute the <strong>Mean Squared Error (MSE)</strong> for:
</p>

<ul style="font-size:20px;">
  <li><strong>x:</strong> input values</li>
  <li><strong>y:</strong> measured / true values</li>
  <li><strong>theta_hat:</strong> estimated slope parameter</li>

<p style="font-size:20px;"><strong>Expected Output</strong></p>

<ul style="font-size:20px;">
  <li><strong>theta_hat = 0.75:</strong> MSE = 9.08</li>
  <li><strong>theta_hat = 1.0:</strong> MSE = 3.00</li>
  <li><strong>theta_hat = 1.5:</strong> MSE = 4.52</li>
</ul>


Мы искусственно создаём данные, потому что у нас пока нет настоящих измерений. Поэтому компьютер сам делает пример данных, чтобы мы могли потренироваться. Потом с помощью MSE пытаемся найти, какой theta_hat лучше всего подходит к этим данным.

```python
np.random.seed(121)   # фиксирует случайность, чтобы каждый раз получались одинаковые данные

theta = 1.2           # задаёт настоящий коэффициент наклона линии
n_samples = 30        # задаёт количество точек данных

x = 10 * np.random.rand(n_samples)   # создаёт 30 случайных значений x от 0 до 10
noise = np.random.randn(n_samples)   # создаёт случайный шум
y = theta * x + noise                # создаёт значения y по формуле y = 1.2 * x + шум
```

```python
from sklearn.metrics import mean_squared_error


def mse1(x, y, theta_hat):

  # Compute the estimated y
  y_hat = theta_hat * x

  # Compute mean squared error
  mse = mean_squared_error(y, y_hat)

  return mse
```

```python
def mse(x, y, theta_hat):

  # Compute the estimated y
  y_hat = theta_hat * x

  # Compute mean squared error
  mse = np.mean((y - y_hat) ** 2)

  return mse
```

```python
theta_hats = [0.75, 1.0, 1.5]

print("MSE #1")
for theta_hat in theta_hats:
  print(f"theta_hat of {theta_hat} has an MSE of {mse(x, y, theta_hat):.2f}")
    
print("MSE #2")
for theta_hat in theta_hats:
  print(f"theta_hat of {theta_hat} has an MSE of {mse1(x, y, theta_hat):.2f}")
```

```python
# @markdown Execute this cell to visualize estimated models

fig, axes = plt.subplots(ncols=3, figsize=(15, 5))

for theta_hat, ax in zip(theta_hats, axes):

  # True data
  ax.scatter(x, y, label='Observed')  # our data scatter plot

  # Compute and plot predictions
  y_hat = theta_hat * x
  ax.plot(x, y_hat, color='r', label='Fit')  # our estimated model

  ax.set(
      title= fr'$\hat{{\theta}}$= {theta_hat}, MSE = {np.mean((y - y_hat)**2):.2f}',
      xlabel='x',
      ylabel='y'
  );

axes[0].legend()
plt.show()

print("Этот код визуализирует три разные модели с разными значениями theta_hat и показывает, насколько хорошо каждая линия подходит к данным.")
```

  Мы выбираем тот вариант <code>theta_hat</code>, у которого

<br>

<p style="font-size:18px;"><strong>Step 1: Prediction Formula</strong></p>

<p style="font-size:15px;">
For each data point, the model predicts:
</p>

<p style="font-size:18px;">
\[
\hat{y}_i = \hat{\theta}x_i
\]
</p>

<br>

<p style="font-size:18px;"><strong>Step 2: Error</strong></p>

<p style="font-size:15px;">
The error is the difference between the true value and predicted value:
</p>

<p style="font-size:18px;">
\[
y_i - \hat{y}_i
\]
</p>

<br>

<p style="font-size:18px;"><strong>Step 3: Mean Squared Error</strong></p>

<p style="font-size:15px;">
MSE squares each error and then takes the average:
</p>

<p style="font-size:18px;">
\[
MSE = \frac{1}{N}\sum_{i=1}^{N}(y_i - \hat{y}_i)^2
\]
</p>

<br>

<p style="font-size:18px;"><strong>Step 4: MSE for This Model</strong></p>

<p style="font-size:15px;">
Since \(\hat{y}_i = \hat{\theta}x_i\), we can write:
</p>

<p style="font-size:18px;">
\[
MSE = \frac{1}{N}\sum_{i=1}^{N}(y_i - \hat{\theta}x_i)^2
\]
</p>

<br>

<p style="font-size:25px;"><strong>Coding Exercise 2: Solve for the Optimal Estimator</strong></p>

<p style="font-size:20px;">
In this exercise, we compute the <strong>optimal theta_hat</strong> using
<strong>least-squares optimization</strong>.
</p>

<ul style="font-size:20px;">
  <li><strong>x:</strong> input values</li>
  <li><strong>y:</strong> measured / true values</li>
  <li><strong>theta_hat:</strong> optimal estimated slope parameter</li>
</ul>

<p style="font-size:20px;">
Instead of trying many different values, we solve for the best
<code>theta_hat</code> directly using the normal equation:
</p>

<p style="font-size:26px; text-align:center;">
  \( \hat{\theta} = \frac{x^T y}{x^T x} \)
</p>

<p style="font-size:20px;"><strong>Expected Output</strong></p>

<ul style="font-size:20px;">
  <li><strong>theta_hat:</strong> approximately 1.21</li>
  <li><strong>MSE:</strong> approximately 1.43</li>
</ul>

```python
def solve_normal_eqn(x, y):
  """Solve the normal equations to produce the value of theta_hat that minimizes MSE.
    Args:
    x (ndarray): An array of shape (samples,) that contains the input values.
    y (ndarray): An array of shape (samples,) that contains the corresponding measurement values to the inputs.
  Returns:
    float: the value for theta_hat arrived from minimizing MSE
  """

  # Compute theta_hat analytically
  theta_hat = np.sum(x * y) / np.sum(x ** 2)
  print("theta_hat =", theta_hat)
    
  theta_hat2 = np.dot(x, y) / np.dot(x, x)
  print("theta_hat2 =", theta_hat2)
    
  X = x.reshape(-1, 1)
  theta_hat3 = np.linalg.lstsq(X, y, rcond=None)[0][0]
  print("theta_hat3 =", theta_hat3)

  return theta_hat


theta_hat = solve_normal_eqn(x, y)
y_hat = theta_hat * x

plot_observed_vs_predicted(x, y, y_hat, theta_hat)
```

```python
# @title Plotting Functions

def plot_observed_vs_predicted(x, y, y_hat, theta_hat):
  """ Plot observed vs predicted data

  Args:
    x (ndarray): observed x values
    y (ndarray): observed y values
    y_hat (ndarray): predicted y values
    theta_hat (ndarray):
  """
  fig, ax = plt.subplots()
  ax.scatter(x, y, label='Observed')  # our data scatter plot
  ax.plot(x, y_hat, color='r', label='Fit')  # our estimated model
  # plot residuals
  ymin = np.minimum(y, y_hat)
  ymax = np.maximum(y, y_hat)
  ax.vlines(x, ymin, ymax, 'g', alpha=0.5, label='Residuals')
  ax.set(
      title=fr"$\hat{{\theta}}$ = {theta_hat:0.2f}, MSE = {np.mean((y - y_hat)**2):.2f}",
      xlabel='x',
      ylabel='y'
  )
  ax.legend()
  plt.show()
```

<p style="font-size:25px;"><strong>Summary</strong></p>

<p style="font-size:20px;">
  Linear least squares regression is an <strong>optimization procedure</strong>
  that can be used for data fitting.
</p>

<ul style="font-size:20px;">
  <li><strong>Task:</strong> predict a value for <code>y</code> given <code>x</code></li>
  <li><strong>Performance measure:</strong> Mean Squared Error, <strong>MSE</strong></li>
  <li><strong>Procedure:</strong> minimize MSE by solving the normal equations</li>
</ul>

<p style="font-size:20px;">
  <strong>Key point:</strong> We fit the model by defining an objective function
  and minimizing it.
</p>

<p style="font-size:25px;"><strong>Least Squares Optimization Derivation</strong></p>

<p style="font-size:20px;">
  We minimize the MSE by setting its derivative with respect to
  <code>theta</code> equal to zero.
</p>

<p style="font-size:20px;">
  The error function is:
</p>

<p style="font-size:24px; text-align:center;">
  \( \frac{1}{N} \sum_{i=1}^{N} (y_i - \theta x_i)^2 \)
</p>

<p style="font-size:20px;">
  After taking the derivative and solving for <code>theta</code>, we get:
</p>

<p style="font-size:24px; text-align:center;">
  \( \hat{\theta} =
  \frac{\sum_{i=1}^{N} x_i y_i}
  {\sum_{i=1}^{N} x_i^2} \)
</p>

<p style="font-size:20px;">
  In vector form:
</p>

<p style="font-size:24px; text-align:center;">
  \( \hat{\theta} =
  \frac{x^T y}{x^T x} \)
</p>

<p style="font-size:20px;">
  This is called solving the <strong>normal equations</strong>.
</p>
